```python
# This is for visualing the graph only

# Importing necessary modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt   
```

```python
# Reading File
Data = pd.read_csv("NVDA.csv")

# Dropping Unnamed columns
Change = {"Unnamed: 11": "a",
        "Unnamed: 12": "b"}
Data.rename(columns=Change, inplace=True)
new_drop = ["a","b"]
Data.drop(new_drop, inplace=True, axis=1)

# Making new file
Year = int(2017)
Month = 6
TVal = Data[Data["Year"].isin([Year])]
LVal = TVal[TVal["Month"].isin([Month])]

# Rounding
Val = LVal.round(decimals = 3)
Val
```

```python
# X axis will remain as Day
X = Val["Day"]

# Y axis changes based on which line is to be shown
Y = Val["Average"]
Z = Val["High"]
W = Val["Low"]
```

```python
# Plotting Lowest, Average and Highest Values in one graph
fig, ax = plt.subplots()
ax.plot(X,W, label = "Lowest")
ax.plot(X,Y, label = "Average")
ax.plot(X,Z, label = "Highest")

# Creating labels to be displayed
ax.set_title(f"Nvidia Stock prices in {Year}")
LabelX = ax.set_xlabel("Date")
LabelY = ax.set_ylabel("Stock prices on Average")
    
# Legend and Showing Graph
plt.legend()
plt.show()
```
