```python
# Required libraries
import numpy as np               # Not used
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import file
df = pd.read_csv("exchange_rates.csv")
df
```

```python
year = 2022
month = 7
cur = "GBP"
```

```python
# Create a dataframe which only consists of data in given currenct, month and year
df = df[df["Year"].isin([year])]
df = df[df["Month"].isin([month])]
Val = df[df["currency"].isin([cur])]

Val
```

```python
# Assign numerical month to actual month for title later
m = ["January","February","March","April","May","June","July","August","September","October","November","December"]
choice = m[month-1]
print(choice)
```

```python
# Making the actual graph
# Initialising X and Y axis
X = Val["Day"]
Y = Val["value"]

# Standard plotting
fig, ax = plt.subplots()
ax.plot(X,Y)

# Label axis and title
LabelX = ax.set_xlabel("Day")
LabelY = ax.set_ylabel("Value")
Title = ax.set_title(f"{cur} Price in {choice} {year}")

```

```python
# Display max and min values
max = Val["value"].max()
min = Val["value"].min()
print("The max value is: ",max)
print("The min value is: ",min)

# Display difference between max and min values
Diff = max - min
print("The largest difference is: ", round(Diff,10))
```
