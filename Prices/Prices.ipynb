```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Data = pd.read_csv("dataset.csv")

drops = ["SNo","timestamp","volume","open","close"]
Data.drop(drops, inplace=True, axis=1)

Year = int(2013)
Month = 5
Currency = "Bitcoin"

TVal = Data[Data["Year"].isin([Year])] 
LVal = TVal[TVal["Month"].isin([Month])]
CVal = LVal[LVal["crypto_name"].isin([Currency])]

Val = CVal.round(decimals = 2)
Val
```

```python
X = Val["Day"]

Y = Val["high"]
W = Val["low"]
```

```python
fig, ax = plt.subplots()
m, b = np.polyfit(X,Y,1)
n, c = np.polyfit(X,W,1)
ax.plot(X,Y, label = "Highest")
ax.plot(X,W, label = "Lowest")

ax.plot(X,((m*X+b)+(n*X+c))/2)

ax.set_title(f"{Currency} prices in {Year}")
LabelX = ax.set_xlabel("Day")
LabelY = ax.set_ylabel("Stock Prices")

plt.legend()
plt.show()
```
