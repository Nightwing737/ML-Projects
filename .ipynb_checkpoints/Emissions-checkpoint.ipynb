```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Data = pd.read_csv("Methane_final.csv")
Val = Data[Data["region"] != "World"]
Val
```

```python
All = pd.unique(Val["region"])
print(All)
```

```python
for i in range(0,All.size):
    region = All[i]
    print(region)
    Val = Data[Data["region"].isin([region])]
    Total = Val["emissions"].sum()
    print(Total)
```
