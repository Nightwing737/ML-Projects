```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('datatraining.csv')
to_drop = ['date']
data.drop(to_drop, inplace = True, axis = 1)
data.rename(columns = {'S.no.':'Number'}, inplace = True)
data
```

```python
Number = data['Number']
Temperature = data['Temperature']
Humidity = data['Humidity']
Light = data['Light']
CO2 = data['CO2']
HumidityRatio = data['HumidityRatio']
Occupancy = data['Occupancy']
```

```python
fig = plt.figure()
myaxes = fig.add_axes([1, 1, 1, 1])
myaxes.plot(Number,Temperature,'r',label = 'Temperature')
myaxes.legend()

fig = plt.figure()
myaxes2 = fig.add_axes([1, 1, 1, 1])
myaxes2.plot(Number,Humidity,'b',label = 'Humidity')
myaxes2.legend()

fig = plt.figure()
myaxes3 = fig.add_axes([1, 1, 1, 1])
myaxes3.plot(Number,Light,'y',label = 'Light')
myaxes3.legend()

fig = plt.figure()
myaxes4 = fig.add_axes([1, 1, 1, 1])
myaxes4.plot(Number,Temperature,'purple',label = 'CO2')
myaxes4.legend()

fig = plt.figure()
myaxes5 = fig.add_axes([1, 1, 1, 1])
myaxes5.plot(Number,HumidityRatio,'g',label = 'HumidityRatio')
myaxes5.legend()

fig = plt.figure()
myaxes6 = fig.add_axes([1,1,1,1])
myaxes6.scatter(Number,Occupancy,s=10, color=(0,0,0),alpha=0.5)
```
